## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
import re
```

```python
weather = pd.read_csv("/Users/celinaagostinho/Desktop/lab-data-cleaning/weather-raw.csv")
weather
```

```python
weather.columns
```

```python
weather = weather.rename(columns={"id" : "weather_station"})
weather
```

```python
weather["month-year"] = weather["month"].astype(str) + "-" + weather["year"].astype(str)
weather
```

```python
# month-year column: how to get day from column name?
# transpose only min and max to column as well as daily values
# put date column closer to beginning: one row will be one date
# check data types
# .fillna
# add id column and generate ids
```

```python

columns = ['weather_station', 'month-year', 'year', 'month', 'element']

weather = pd.melt(weather, id_vars = columns)
weather
```

```python
weather.head(10)
```

```python
weather.shape
```

```python
weather["date"] = weather["variable"].astype(str) + "-" + weather["month-year"].astype(str)
weather
```

```python
weather = weather.rename(columns={"value" : "reading"})
```

```python
date_re = re.compile(r'[a-z]')

def correct_date(x):
  return date_re.sub('',x)
```

```python
weather["date"] = weather["date"].astype(str).apply(correct_date)
```

```python
weather
```

```python
#Make a list to drop
drop_cols = ["variable"]   
weather = weather.drop(drop_cols, axis=1)
```

```python

weather["tmax"] = weather["reading"].where(weather["element"]=="tmax")

#job_title["jobTitle"]=job_title["jobTitle"].where (job_title["jobTitle"]!="" ,"Not Disclosed")
```

```python
weather.tail(10)
```

```python
weather["tmin"] = weather["reading"].where(weather["element"]=="tmin")
```

```python
#Make another list to drop
#I'm keeping year and month because it can be useful for some operations
drop_cols = ["reading", "element", "month-year"]  
weather = weather.drop(drop_cols, axis=1)
```

```python
weather.tail(10)
```

```python
# remove duplicates

weather = weather.drop_duplicates()
weather.tail()
```

```python
#I want to have tmin and tmax in the same row, one row per date
#groupby creates multiindex

weather = weather.groupby(["weather_station", "month", "year", "date"]).sum()
weather.tail(10)
```

```python
weather.shape
```

```python
#does not remove 0.0

weather = weather.fillna("NA")
```

```python
weather["tmax"].dtypes
```

```python
#Use a where condition
#As far as I can tell, there were no temp readings of 0.0, or of negative values
weather["tmax"] = weather["tmax"].where(weather["tmax"]!=0.0 ,"NA")
```

```python
weather.tail(10)
```

```python
weather["tmin"] = weather["tmin"].where(weather["tmin"]!=0.0 ,"NA")
```
